```python
# Import dependencies
from bs4 import BeautifulSoup
from splinter import Browser
import pandas as pd
import requests
```

```python
## NASA Mars News
```

```python
# URL of page to be scraped
url = 'https://mars.nasa.gov/news/?page=0&per_page=40&order=publish_date+desc%2Ccreated_at+desc&search=&category=19%2C165%2C184%2C204&blank_scope=Latest'
# Retrieve page with the requests module
response = requests.get(url)
# Create BeautifulSoup object; parse with 'html.parser'
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Examine the results, then determine element that contains sought info
print(soup.prettify())
```

```python
# Extract needed data.
news_title = soup.find("div",class_="content_title").text
news_p = soup.find('div', class_="rollover_description_inner").text

print(news_title)
print(news_p)
```

```python
# JPL Mars Space Images
```

```python
# https://splinter.readthedocs.io/en/latest/drivers/chrome.html
!which chromedriver
```

```python
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Set URL.
jpl_url = 'https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars/'
browser.visit(jpl_url)
```

```python
# Create BS object
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
# create link to featured image. 
featured_img_base = "https://www.jpl.nasa.gov"
featured_img_url_raw = soup.find("div", class_="carousel_items").find("article")["style"]
featured_img_url_raw
```

```python
# Remove all but the url, concatenate the two. 
featured_img_url = featured_img_url_raw.split("'")[1]
featured_img_url = featured_img_base + featured_img_url
featured_img_url
```

```python
# Mars Weather

# Visit the Mars Weather twitter account.
twitter_url = 'https://twitter.com/marswxreport?lang=en'
browser.visit(twitter_url)
```

```python
# Scrape data and look for relevant info.
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
print(soup.prettify())
```

```python
# Scrape most recent tweet.
mars_weather = soup.find("div",class_="js-tweet-text-container").text
print(mars_weather)
```

```python
# Mars Facts

facts_url = "http://space-facts.com/mars/"
browser.visit(facts_url)
```

```python
# Scrape the site for relevant data.
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
print(soup.prettify())
```

```python
# Use pandas to scrape tabulated data
tables = pd.read_html(facts_url)
tables
```

```python
facts_df = tables[0]
facts_df.columns = ["Fact","Value"]
facts_df
```

```python
# Change the index for better readability. 
facts_df = facts_df.set_index("Fact")
facts_df
```

```python
# Convert table to HTML format. 
facts_html_table = facts_df.to_html()
facts_html_table = facts_html_table.replace('\n', '')
facts_html_table
```
